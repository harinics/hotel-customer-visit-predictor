```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load retail dataset
data = pd.read_csv("C:\\Users\\Harini CS\\Downloads\\projectf\\data\\enhanced_retail_customer_visits.csv")

# Basic Info
print(" Dataset Info:")
print(data.info())

print("\n First 5 Rows:")
print(data.head())

print("\n Missing Values:")
print(data.isnull().sum())

print("\n Summary Statistics:")
print(data.describe(include='all'))

print("\n Unique Values Per Column:")
for col in data.columns:
    print(f"{col}: {data[col].nunique()}")

# EDA Plots
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Visited')
plt.title("Count of Customer Visits")
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Gender', hue='Visited')
plt.title("Gender vs Visited")
plt.show()

plt.figure(figsize=(8, 4))
sns.countplot(data=data, x='Day', hue='Visited')
plt.title("Day of Visit vs Visited")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Weather', hue='Visited')
plt.title("Weather vs Visited")
plt.show()

# Label Encoding
from sklearn.preprocessing import LabelEncoder

df = data.copy()
df.columns = df.columns.str.strip()  # Fix hidden spaces

for col in df.select_dtypes(include='object'):
    df[col] = LabelEncoder().fit_transform(df[col])

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Prepare Features & Target
X = df.drop(columns=['Visited', 'Date', 'Customer_ID'])
y = df['Visited']

# Train-Test Split
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn import svm, metrics

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# SVM Model with GridSearchCV
clf = svm.SVC(class_weight='balanced', random_state=42)
param_grid = {
    'kernel': ['linear', 'rbf', 'poly'],
    'C': [0.01, 0.1, 1, 10, 100],
    'gamma': ['scale', 0.001, 0.01, 0.1, 1],
    'degree': [2, 3, 4]
}

grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='f1', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Best Model
best_clf = grid_search.best_estimator_
best_clf.fit(X_train_scaled, y_train)

# Predictions
y_pred = best_clf.predict(X_test_scaled)

# Evaluation
print("\nModel Evaluation:")
print(f"Accuracy: {metrics.accuracy_score(y_test, y_pred):.2f}")
print(f"Precision: {metrics.precision_score(y_test, y_pred):.2f}")
print(f"Recall: {metrics.recall_score(y_test, y_pred):.2f}")
print(f"F1-Score: {metrics.f1_score(y_test, y_pred):.2f}")

print("\nClassification Report:")
print(metrics.classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
#knn
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

data = pd.read_csv("C:\\Users\\Harini CS\\Downloads\\projectf\\data\\hospital_patient_visits_enhanced.csv")

data = data.dropna()

X = data.drop(columns=['Visited', 'Day', 'Holiday', 'Age', 'Patient_ID', 'Date', 'Age_Group'])  
y = data['Visited'] 

label_enc = LabelEncoder()

X['Chronic_Condition'] = label_enc.fit_transform(X['Chronic_Condition'])
X['First_Visit_Date'] = label_enc.fit_transform(X['First_Visit_Date'])
X['Disease'] = label_enc.fit_transform(X['Disease'])
X['Visit_Count'] = label_enc.fit_transform(X['Visit_Count'])
X['Visit_Frequency'] = label_enc.fit_transform(X['Visit_Frequency'])
X['Visit_Dates'] = label_enc.fit_transform(X['Visit_Dates'])
X['Time_Between_Visits_Avg'] = label_enc.fit_transform(X['Time_Between_Visits_Avg'])
X['Patient_Tenure_Days'] = label_enc.fit_transform(X['Patient_Tenure_Days'])
X['Season_of_Visit'] = label_enc.fit_transform(X['Season_of_Visit'])
X['Weekend_Visit'] = label_enc.fit_transform(X['Weekend_Visit'])
X['Gender'] = label_enc.fit_transform(X['Gender'])
X['Is_Chronic'] = label_enc.fit_transform(X['Is_Chronic'])
X['Gender_Age_Group'] = label_enc.fit_transform(X['Gender_Age_Group'])
X['Large_Visit_Gaps'] = label_enc.fit_transform(X['Large_Visit_Gaps'])
X['Visits_Last_3_Months'] = label_enc.fit_transform(X['Visits_Last_3_Months'])
X['Visit_in_Last_30_Days'] = label_enc.fit_transform(X['Visit_in_Last_30_Days'])
X['Previous_Visit_Day_Type'] = label_enc.fit_transform(X['Previous_Visit_Day_Type'])
X['Visit_Trend'] = label_enc.fit_transform(X['Visit_Trend'])
X['Time_Since_First_Visit'] = label_enc.fit_transform(X['Time_Since_First_Visit'])
X['Holiday_Season_Effect'] = label_enc.fit_transform(X['Holiday_Season_Effect'])
X['Weekend_Visit'] = label_enc.fit_transform(X['Weekend_Visit'])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier()
param_grid = {
    'n_neighbors': [3, 5, 7, 10, 15], 
    'metric': ['euclidean', 'manhattan', 'minkowski']  
}

grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='f1', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

best_knn = grid_search.best_estimator_

best_knn.fit(X_train_scaled, y_train)

y_pred = best_knn.predict(X_test_scaled)

print("\nModel Evaluation:")
print(f"Accuracy: {metrics.accuracy_score(y_test, y_pred):.2f}")
print(f"Precision: {metrics.precision_score(y_test, y_pred, pos_label='Yes'):.2f}") 
print(f"Recall: {metrics.recall_score(y_test, y_pred, pos_label='Yes'):.2f}") 
print(f"F1-Score: {metrics.f1_score(y_test, y_pred, pos_label='Yes'):.2f}") 

print("\nClassification Report:")
print(metrics.classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(metrics.confusion_matrix(y_test, y_pred))
```
